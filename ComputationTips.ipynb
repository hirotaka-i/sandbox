## Using jupyter notebook on server

1. log-in to the server 
2. load tmux so that the sinteractive keep active even the pipe breaks
3. open interactive node
4. run jupyter notebook at your workplace with port option (open port). 
5. connect the port and log-in to the sever
6. tunnel to the node you are running the jupyter-notebook

If you lose the connection, you can connect to the jupyter notebook by doing only 5 and 6. 
The tunneling terminal doesn't have access to the internet. So you cannot install a new package or push git. You have two options to do this. 1. stop the jupyter notebook on the original terminal and install packages. 2. (For git,) just push a new version through your local network (my recommendation).

```
ssh xxxx@xxxxxx.xxx.xxx
# INPUT PASSWORD
module load tmux
tmux  # sometime you may want to do "tmux ls" to check the available windows. close unnecessary ones.
module load python/3.5
module load R # optional. To use R kernel on notebook
cd $YourWorkSpace  # Workplace you would like to start jupyter notebook
jupyter notebook --port 2222 --no-browser # portnumber is your preferred 4-digit numbers

# From another terminal
ssh -L 2222:localhost:2222 xxxxxx.xxx.xxx # Don't need username
# INPUT PASSWORD
ssh -L 2222:localhost:2222 cn3333 # this is the node your interactive session is running.
```


## Git
Very simple way to use git. I found it much easier to do this on the local network, not on the server. 
```git
cd $YourWorkingDirectory
# only for the first time
git init 
# Check status
git status
# add & commit
git add $YourScriptFile
git commit -m 'Whatever'
# See update log 
git log
# Push to the github. DON'T push things contains sensitive data!!
## First create git account and create repository. Only one-time
git remote add origin https://github.com/user/repository_name.git
## push commitment
git push -u origin master
```

## Some bash essentials for bio-infomatics
#### Create a variable and interpretations of "echo" command
```
%%bash
TEST=somthing
echo "$TEST"
echo $TEST
echo $TEST1
echo "$TEST"1
echo '$TEST'
echo ls
echo `ls`
echo $(ls)
```

#### For-loop
```
%%bash
for i in "TEST1" "TEST2";do
    for j in {1..5};do 
        echo "$j of $i"
    done
    echo "next loop"
done

# Output to a file.
for i in "TEST1" "TEST2";do;
    for j in {1..5};do;
        echo "$j of $i" >> sample.txt;
    done;
    echo "next loop" >> sample.txt
done
```

#### Running Rscript from Terminal
```
%%bash
echo '
args <- commandArgs(trailingOnly = TRUE)
x1 = INPUT1
x2 = INPUT2
x3 = INPUT3
y = x1 + x2 + x3
write.table(y, "example.csv")
' > example.R
Rscript --vanilla example.R INPUT1 INPUT2 INPUT3 >example.Rout 2>&1 # error and output together.
```

#### Use Lang=C for sorting!
"sort" should be always used with LANG=C for "join". If you like to keep the first row, use awk too. 
```
# ordinary sort with LANG=C and add the header
LANG=C sort VariantsPosName.txt | sed -i'1 iPos\tName'
# Read csv and keep the first raw as it is and join with the above file by its position
awk 'BEGIN{OFS=","}NR==1{print;next}{print|"LANG=C sort"}' ANALYSIS.txt |
    LANG=C join --header -t$'\t' VariantsPosName.txt - > NameAnalysis.txt
```
